```
import sklearn;
import numpy as np;
import pandas as pd;
from sklearn.preprocessing import LabelEncoder;
from sklearn.preprocessing import StandardScaler;
```

```
data = {
'size':['S', 'M', 'L', 'XL', 'XXL', 'M', 'M', 'XXL', 'XL', 'S'],
'color':['green', 'red', 'blue', 'white', 'dark', 'yellow', 'grey', 'white', 'dark', 'yellow'],
'gender':['female','male', 'male', 'male', 'female', 'female', 'male', 'female', 'male', 'female',],
'price':[110.0, 120.0, 130.0, 140.0, 150.0, 160.0, 170.0, 180.0, 190.0, 200.0],
'weight':[250, 275, 300, 325, 350, 375, 400, 425, 450, 475],
'bought':['yes', 'no', 'no','yes','yes','yes','no','no','yes','no']
}

clothes = pd.DataFrame(data=data)
print(clothes)
clothesCopy = clothes.copy();

clothesCopy['weight'] = clothesCopy['weight'].astype('float64');

for col in ['size', 'color', 'gender', 'bought' ]:
  clothesCopy[col] = clothesCopy[col].astype('category');
```

```
floats = clothesCopy.describe().T
categories = clothesCopy.describe(include=['category']).T
print(floats)
print(categories)
```

```
le = LabelEncoder()
le.fit_transform(clothesCopy['bought'])
clothesCopy['bought'] = le.fit_transform(clothesCopy['bought'])
clothesCopy
```

```
pd.get_dummies(data=clothesCopy, drop_first=True)
```

```
scaler = StandardScaler();
clothesCopy[['price', 'weight']] = scaler.fit_transform(clothesCopy[['price', 'weight']])
```

```
print(clothes)
print(clothesCopy)
```
