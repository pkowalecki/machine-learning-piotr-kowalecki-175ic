```
import numpy as np
import pandas as pd
import sklearn 
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = {
'size':['S', 'M', 'L', 'XL', 'XXL', 'M',np.nan, 'M', 'XXL', 'XL', 'S', np.nan],
'color':['green', 'red', 'blue', 'white',np.nan, 'dark', 'yellow', 'grey', 'white', 'dark', 'yellow', np.nan],
'gender':['female','male', 'male', 'male',np.nan, 'female', 'female', 'male', 'female', 'male', 'female', np.nan],
'price':[110.0, 120.0, 130.0, 140.0,np.nan, 150.0, 160.0, 170.0, 180.0, 190.0, 200.0, np.nan],
'weight':[250, 275, 300, 325, 350,np.nan, 375, 400, 425, 450, 475, np.nan],
'bought':['yes', 'no', 'no','yes',np.nan,'yes','yes','no','no','yes','no', np.nan]
}

df_raw = pd.DataFrame(data=data)
df = df_raw.copy()
df
```

```
#1.	pomiń dane, które mają 3 lub więcej pustych wartości
df.dropna(thresh=len(df.keys())-2, inplace=True)
df
```

```
#2.	wypełnij puste miejsca w kolumnie price wartością średnią
df['price'].fillna(df['price'].mean(), inplace=True)
df
```

```
#3.	wypełnij puste miejsca w kolumnie weight za pomocą interpolacji
df['weight'].fillna(df['weight'].interpolate(), inplace=True)
df
```

```
#4.	wypełnij puste miejsca w kolumnie size, color i gender wartościami stałymi (np. unisex, black, male)
si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='TooBig')
df['size'] = si.fit_transform(df[['size']])

si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='black')
df['color'] = si.fit_transform(df[['color']])

si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='male')
df['gender'] = si.fit_transform(df[['gender']])

si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='yes')
df['bought'] = si.fit_transform(df[['bought']])

df
```

```
#5.	dodaj do słownika kolumnę cost, która reprezentuje cenę jako jedną z klas:
#cheap, normal_price, expensive; podziału ceny na te klasy dokonaj automatycznie (argument bins metody pandas.cut)

df['cost'] = pd.cut(df['price'], bins=3, labels=('cheap', 'normal_price', 'expensive'))
df
```

```
#6.	dostosuj dane do modelu tak, by przedstawiały wartość binarną (np. bought: [‘yes’, ‘no’, ‘yes’] na bought_yes: [1,0,1])

df['bought_yes'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
df
```

```
#7.	podziel słownik na dane wejściowe i wartość wyjściową, a następnie podziel na zbiór uczący i testujący
from sklearn.model_selection import train_test_split

df_in=df[['price', 'cost', ]]
df_out=df[['bought_yes']]

x_train, x_test, y_train, y_test = train_test_split(df_in, df_out, train_size=0.2)

x_train, x_test, y_train, y_test
```
