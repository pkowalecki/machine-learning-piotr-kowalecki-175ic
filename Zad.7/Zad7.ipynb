```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

```
blobs,_ = make_blobs(n_samples=1000, centers=4,random_state=96,cluster_std=1.5)
df = pd.DataFrame(blobs,columns=['x1','x2'])
px.scatter(df,'x1','x2')
```

```
cluster = DBSCAN(eps=1.1,min_samples=4)
cluster.fit(df)
df['cluster'] = cluster.labels_
px.scatter(df,'x1','x2','cluster',template='plotly_dark',color_continuous_midpoint=0)
```

```
k_means = KMeans(n_clusters=4,random_state=0)
k_means.fit(df)
df['k_means'] = k_means.labels_
px.scatter(df,'x1','x2','k_means', template='plotly_dark')
```

```
pred = k_means.fit_predict(df)
result = pd.DataFrame([df.index,pred]).T
print(f'KMeans')
print(result[1].value_counts())
print(f'\n DBSCAN')
print(df['cluster'].value_counts())
```
